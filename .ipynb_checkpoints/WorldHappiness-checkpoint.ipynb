### Importing Required Modules

```python
# Importing modules
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

### Read CSV Data and Store in DataFrame

```python
# Read file and store it in DataFrame
orig_data_2018 = pd.read_csv("Resource/2018.csv")
orig_data_2019 = pd.read_csv("Resource/2019.csv")
orig_data_2020 = pd.read_csv("Resource/2020.csv")



# Get only required columns from dataframe
data_2018 = orig_data_2018[["Country or region", "Score", "GDP per capita", "Healthy life expectancy", 
                            "Perceptions of corruption"]]

# Rename Columns so they are uniform across all years
data_2018 = data_2018.rename(columns={"Country or region" : "Country",
                                      "Healthy life expectancy" : "Health",
                                      "GDP per capita" : "GDP_per_Capita",
                                      "Perceptions of corruption" : "Corruption"})

# Get only required columns from dataframe
data_2019 = orig_data_2019[["Country or region", "Score", "GDP per capita", "Healthy life expectancy",
                            "Perceptions of corruption"]]

# Rename Columns so they are uniform across all years
data_2019 = data_2019.rename(columns={"Country or region" : "Country",
                                      "Healthy life expectancy" : "Health",
                                      "GDP per capita" : "GDP_per_Capita",
                                      "Perceptions of corruption" : "Corruption"
                                      })

# Get only required columns from dataframe
data_2020 = orig_data_2020[["Country name", "Regional indicator", "Ladder score", "Logged GDP per capita",
                            "Healthy life expectancy", "Perceptions of corruption"]]

# Rename Columns so they are uniform across all years
data_2020 = data_2020.rename(columns = {"Country name" : "Country",
                                        "Regional indicator" : "Region",
                                        "Ladder score" : "Score",
                                        "Logged GDP per capita" : "GDP_per_Capita",
                                        "Healthy life expectancy" : "Health",
                                        "Perceptions of corruption" : "Corruption"})

```

### Data Cleanup
* Merge all three DF into one DF
* Get average of Score, GDP, Health, Corruption for all years
* Recategorize Region 

```python
# Merge all three DF to one by adding suffixes
merged_data = pd.merge(pd.merge(data_2020,data_2019,on = "Country", how="inner", suffixes=[None,'_2019']),
                       data_2018,on = "Country", how="inner",suffixes=['_2020','_2018'])

# Display merged Data Frame
merged_data
```

```python
# Calculate average of score, GDP, Heath, Corruption over year and add it to column as Avgscore, AvgGDP, AvgHealth, AvgCorruption
# Create a dataframe of that
Average_data = pd.DataFrame({"Country" : merged_data["Country"],
                             "Region" : merged_data["Region"],
                             "AvgScore" : merged_data[["Score_2020", "Score_2019", "Score_2018"]].mean(axis=1),
                             "AvgGDP" : merged_data[["GDP_per_Capita_2020", "GDP_per_Capita_2019", "GDP_per_Capita_2018"]].mean(axis=1),
                             "AvgHealth" : merged_data[["Health_2020", "Health_2019", "Health_2018"]].mean(axis=1),
                             "AvgCorruption" : merged_data[["Corruption_2020", "Corruption_2019", "Corruption_2018"]].mean(axis=1)
                            })

```

```python
Average_data["Region"].unique()

# Combine (Rename) all regions which has divided into multiple regions for example rename regions 
# "South Asia" & "Southeast Asia" as "Asia"
Average_data= Average_data.replace({"Region" : { "Central and Eastern Europe" : "Europe", 
                                                 "East Asia" : "Asia",
                                                 "Middle East and North Africa" : "Africa",
                                                 "South Asia" : "Asia",
                                                 "Southeast Asia" : "Asia",
                                                 "Sub-Saharan Africa" : "Africa",
                                                 "Western Europe" : "Europe",
                                                 "Commonwealth of Independent States" : "Commonwealth",
                                                 "North America and ANZ" : "N. America",
                                                 "Latin America and Caribbean" : "S. America"
                                                }
                                   })

Average_data["Region"].unique()
```

```python
# Display final Data Frame
Average_data
```

### Question 1 : How does happiness relates on different factors : GDP, Health & Corruption

### Analysis:


```python
# Defined method to calculate linear regression, parameters as below:
# y_values = series/column values to plot on y-axis
# ylabel = string to display as y-label
# image_name = .png file name to save plot as image
# eqplace = location to place equation on plot

def score_linear(y_values, ylabel, image_name, eqplace):
    
    x_values = Average_data["AvgScore"]
    
    # Get the linear equation
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    # Plot the scatter plot
    plt.scatter(x_values,y_values, edgecolors="black", facecolors="purple")
    plt.plot(x_values,regress_values,"r-")
    
    # Annotate the line_equation on graph
    plt.annotate(line_eq,eqplace,fontsize=15,color="blue",fontweight="bold")
    
    # Set label for x & y axis
    plt.xlabel("Avg Happiness Score")
    plt.ylabel(ylabel)
    plt.xlim(2,8)
    
    # Print r-value
    print(f"The r-value is: {rvalue}")
    
    # Save plot image
    plt.savefig(f"Plot_Images/{image_name}")
    
    # Display Plot
    plt.show()
```

```python
# Find out the corelation between Happiness score and GDP:
score_linear(Average_data["AvgGDP"], "Avg GDP", "score_gdp_regression.png", (3, 4.5))
```

```python
# Find out the corelation between Happiness score and Health:
score_linear(Average_data["AvgHealth"], "Avg Health", "score_health_regression.png", (3, 26))
```

```python
# Find out the corelation between Happiness score and Health:
score_linear(Average_data["AvgCorruption"], "Avg Corruption", "score_corruption_regression.png", (2.5, 0.38))
```

### Question 4 : Are there any similarities in demographics in top 5 and Bottom 5 countries?

### Analysis:

#### Top 5 countries

```python
# Sort the data frame by AvgScore in descending order and get the top 5 countries.
top_5 = Average_data.sort_values(by=["AvgScore"], ascending=False).head(5)

# Display top 5 countries
top_5
```

```python
# x-locations for the graph
ind = np.arange(5) 

# Set the figure size
plt.figure(figsize=(10,7))

# Plot the stacked bar graph 
p1 = plt.bar(ind, top_5["AvgHealth"], color='turquoise', edgecolor='white', width=0.5)
p2 = plt.bar(ind, top_5["AvgGDP"], color='goldenrod', edgecolor='white', width=0.5, bottom = top_5["AvgHealth"])
p3 = plt.bar(ind, top_5["AvgCorruption"], color='purple', edgecolor='white', width=0.5,
             bottom = top_5["AvgHealth"] + top_5["AvgGDP"])

# Set the x & yticks
plt.xticks(ind, top_5["Country"], rotation=90)
plt.yticks(np.arange(0, 50, 5))

# Add legend for the plot
plt.legend((p1[0], p2[0], p3[0]), ('AvgHealth', 'AvgGDP', 'AvgCorruption'))

# Save plot image
plt.savefig("Plot_Images/Top5_Stacked_Bar.png")

# Show plot 
plt.show()
```

#### Bottom 5 countries

```python
# Sort the data frame by AvgScore in descending order and get the bottom 5 countries.
bottom_5 = Average_data.sort_values(by=["AvgScore"], ascending=False).tail(5)

# Display top 5 countries
bottom_5
```

```python
# x-locations for the graph
ind = np.arange(5) 

# Set the figure size
plt.figure(figsize=(10,7))

# Plot the stacked bar graph 
p1 = plt.bar(ind, bottom_5["AvgHealth"], color='turquoise', edgecolor='white', width=0.5)
p2 = plt.bar(ind, bottom_5["AvgGDP"], color='goldenrod', edgecolor='white', width=0.5, bottom = bottom_5["AvgHealth"])
p3 = plt.bar(ind, bottom_5["AvgCorruption"], color='purple', edgecolor='white', width=0.5,
             bottom = bottom_5["AvgHealth"] + bottom_5["AvgGDP"])

# Set the x & yticks
plt.xticks(ind, top_5["Country"], rotation=90)
plt.yticks(np.arange(0, 50, 5))

# Add legend for the plot
plt.legend((p1[0], p2[0], p3[0]), ('AvgHealth', 'AvgGDP', 'AvgCorruption'))

# Save plot image
plt.savefig("Plot_Images/Bottom5_Stacked_Bar.png")

# Show plot 
plt.show()
```

### Question 5 : Which is the Happiest Region?

### Analysis:

```python
# Using Matplotlib

# Create a groupby object by region
region_stat = Average_data.groupby("Region")

# Get the average score by region
avg_region_score = region_stat["AvgScore"].mean()

# Get the regions name from index values
regions = list(avg_region_score.index.values)

#Plot the graph
plt.pie(avg_region_score, labels = regions,  autopct="%1.1f%%", shadow=True, startangle=140, rotatelabels =False)

# Set the font size, title & ylabel
plt.rcParams['font.size'] = 16
plt.title("Happiness score by Region", pad=30)
plt.ylabel("Average Score")
plt.axis("equal")

# Save plot image
plt.savefig("Plot_Images/Region_score_pie.png")

# Show plot 
plt.show()
```

```python
# Bar Plot:
region_bar_plot = Average_data.groupby("Region")["AvgScore"].mean().plot(kind='bar', color=tuple(['b', 'g', 'r', 'c', 'm', 'y',]),
                                                       ylabel = "Happiness score", ylim=(0,8), 
                                                       title = "Happiness Score by Region")

# Save plot
region_bar_plot.get_figure().savefig("Plot_Images/Region_score_bar.png", bbox_inches="tight")
```
